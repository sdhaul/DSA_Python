# LIBRARIES

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split, cross_val_score
```

# GENERATING SAMPLE DATA

```python
# Random data
np.random.seed(0)
X = 2*np.random.rand(100,1)
y = 3*X + np.random.rand(100,1)
```

# SPLITTING DATA INTO TRAINING AND TESTING SPLITS:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2 ,random_state=0)
```

# EXECUTING AND VISUALISING POLYNOMIAL REGRESSION

```python
# degree of polynomial
degree=5
poly_features = PolynomialFeatures(degree=degree)
X_train_poly = poly_features.fit_transform(X_train) # ****

# Fit the linear regression model
poly_reg = LinearRegression() # ****
poly_reg.fit(X_train_poly, y_train) # ****

plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, poly_reg.predict(poly_features.transform(X)), color='red' ) # ****
plt.title('Polynomial Regression (Degree {})'.format(degree)) # ****
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

# K-FOLD CROSS VALIDATION

```python
# Perform k fold cross validation
cv_scores = cross_val_score(poly_reg, X_train_poly, y_train, cv=5) # ****
print("Mean CV Score: ", np.mean(cv_scores))
```
